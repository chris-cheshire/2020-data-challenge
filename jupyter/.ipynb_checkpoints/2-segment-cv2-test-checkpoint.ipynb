```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
frames_path = '../data/frames/C59_L1_t0'
```

```python
img_orig = cv2.imread(os.path.join(frames_path,'frame-1.jpg'),0)
img = cv2.medianBlur(img_orig,7)

ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Blurred Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img_orig, img, th1, th2, th3]

for i in range(5):
    plt.figure(figsize=(50,50))
    plt.subplot(5,1,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
